### Since the data didnt have any date fields, i will be adding an extra Date column

```python
months_map = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
def mapper(month):
    return months_map[month]
unemployment['month'] = unemployment['month'].apply(mapper)
```

```python
unemployment.head(12)
```

```python
unemployment.columns
```

```python
from datetime import timedelta
```

```python
unemployment['Date'] = pd.to_datetime(unemployment.assign(Day=1).loc[:, ['year','month','Day']])
```

```python
unemployment.head(12)
```

```python
unemployment.shape
```

```python
unemployment = unemployment.set_index('Date')
```

### Unemployment rate data is per month so i am trying to replicate those values for that whole month respectively.


### To do that, first i will be creating a column with years 2010-18 allong with all the days for each month

```python
import datetime

date = pd.Series(pd.date_range('2010-1-1','2018-12-31'))
date_df = pd.DataFrame(date)
date_df.head()
```

```python
date_df['Month'] = date_df[0].dt.month
date_df['Year'] = date_df[0].dt.year
date_df.head()
```

```python
unemployment_with_days = pd.merge(date_df, unemployment, how = 'left', left_on = ['Month','Year'], right_on = ['month','year'])
```

```python
unemployment_with_days.head()
```

```python
unemployment_with_days.rename(columns={unemployment_with_days.columns[0]:'Date'},inplace=True)
unemployment_with_days.head()
```

```python
unemployment_with_days.shape
```

### Keeping only the usefull features and droping the rest

```python
unemployment_with_days = unemployment_with_days.drop(['Month', 'Year', 'month', 'year'], axis=1)
```
